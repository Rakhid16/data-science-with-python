## While Loop

### With the while loop we can execute a set of statements as long as a condition is true.

```python
error = 50.00
while error > 1 :
    error = error / 4
    print(error)
```

```python
error = 50.0
while error > 1: # BAHAYA ALWAYS TRUE INFINITY LOOP 
    # error = error / 4
    print(error)
```

## For Loop

```python
keluarga = [1.73, 1.68, 1.71, 1.89]
print(keluarga)
```

```python
keluarga = [1.73, 1.68, 1.71, 1.89]
print(keluarga[0])
print(keluarga[1])
print(keluarga[2])
print(keluarga[3])
```

```python
keluarga = [1.73, 1.68, 1.71, 1.89]
for tinggi in keluarga:
    print(tinggi)
```

### enumerate 
Metode Enumerate () menambahkan penghitung ke iterable dan mengembalikannya dalam bentuk objek enumerasi. 
In the example above, iterable is an iterable type object like a list, tuple, or string.

start is an optional parameter that tells the function which index to use for the first item of the iterable.
enumerate(iterable=,start=)

```python
cars = ['kia', 'audi', 'bmw']
for car in enumerate(cars):
  print(car)
```

```python
cars = ['kia', 'audi', 'bmw']
for car in cars:
  print(car)
```

```python
#cars = ['kia', 'audi', 'bmw']
print(list(enumerate(cars, start = 1)))
```

```python
for index, tinggi in enumerate(keluarga,start=0): #deafult start=0
    print("Index " + str(index) + ": " + str(tinggi))
```

```python
for huruf in 'keluarga':
    print(list(huruf.capitalize())) #coba tanpa list
```

## Loop pada dictionary

```python
negara = { "afghanistan":30.55,
"albania":2.77,
"algeria":39.21 }
for key, value in negara.items():
    print(key + " -- " + str(value))
```

## NumPy Arrays

```python
#numpy arrays
import numpy as np
np_tinggi = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_berat = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
bmi = np_berat / np_tinggi ** 2
print(bmi)
for val in bmi:
    print(val)
```

```python
import numpy as np
np_tinggi = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_berat = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
d_2 = np.array([np_tinggi, np_berat])
print(d_2)
for val in d_2:
    print(val)
```

```python
import numpy as np
np_tinggi = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_berat = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
d_2 = np.array([np_tinggi, np_berat])
for val in np.nditer(d_2):
    print(val)
```
