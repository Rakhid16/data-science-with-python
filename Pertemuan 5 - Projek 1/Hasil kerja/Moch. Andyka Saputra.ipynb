# Join

```python
cursor = koneksi.cursor()
sql = "SELECT \
  customers.customerName AS user, \
  products.productLine AS fav \
  FROM customers\
  INNER JOIN products ON customers.country = products.buyPrice"
cursor.execute(sql)
myresult = cursor.fetchall()
for x in myresult:
  print(x)
```
